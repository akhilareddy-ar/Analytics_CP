```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
train_df = pd.read_csv("../data/train.csv")

# Display basic info
print("Dataset Info:")
print(train_df.info())

# Display basic statistics
print("\nBasic Statistics:")
print(train_df.describe())

# Check for missing values
print("\nMissing Values:")
print(train_df.isnull().sum())

# Check column names
print("\nColumns in dataset:", train_df.columns)

# Ensure column names are stripped of spaces
train_df.columns = train_df.columns.str.strip()

# Visualizing the distribution of genres (using 'Class' column as genre)
if 'Class' in train_df.columns:
    plt.figure(figsize=(10,5))
    sns.countplot(x='Class', data=train_df, order=train_df['Class'].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Genre Distribution")
    plt.show()
else:
    print("Error: 'Class' column not found in dataset.")

# Exclude non-numeric columns before computing correlation
numeric_df = train_df.select_dtypes(include=[np.number])

# Correlation heatmap
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```
